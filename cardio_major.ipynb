```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the file

```python
file_path = "D:\Corizo class\Assignment\Minor\cardio_train.csv"
```

```python
df = pd.read_csv(file_path, delimiter=";")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Handling outliers by capping extreme values

# Define reasonable limits for blood pressure (ap_hi, ap_lo)
df = df[(df['ap_hi'] > 50) & (df['ap_hi'] < 200)]  # Typical systolic range
df = df[(df['ap_lo'] > 30) & (df['ap_lo'] < 150)]  # Typical diastolic range

# Define reasonable limits for height and weight
df = df[(df['height'] > 100) & (df['height'] < 250)]  # Height in cm
df = df[(df['weight'] > 30) & (df['weight'] < 200)]  # Weight in kg

# Rechecking the data after outlier removal
df.describe()
```

```python
# Visualization: Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['age'] / 365, bins=30, kde=True, color='blue')  # Age converted to years
plt.xlabel('Age (Years)')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

# Visualization: Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=df, palette='coolwarm')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()

# Visualization: Cholesterol Levels
plt.figure(figsize=(6, 4))
sns.countplot(x='cholesterol', data=df, palette='viridis')
plt.xlabel('Cholesterol Levels')
plt.ylabel('Count')
plt.title('Cholesterol Level Distribution')
plt.show()

# Blood Pressure Distribution - Histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['ap_hi'], bins=30, kde=True, color='red')
plt.xlabel('Systolic Blood Pressure (ap_hi)')
plt.ylabel('Count')
plt.title('Systolic Blood Pressure Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['ap_lo'], bins=30, kde=True, color='blue')
plt.xlabel('Diastolic Blood Pressure (ap_lo)')
plt.ylabel('Count')
plt.title('Diastolic Blood Pressure Distribution')

plt.tight_layout()
plt.show()
```

Seperating x and y

```python
x = df.iloc[:,1:12]
x
```

```python
y = df.iloc[:,12:]
y
```

Spliting the train and test data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.8)
```

# SVC

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(xtrain,ytrain)
ans_svc = svc_model.predict(xtest)
```

```python
acc_svc = svc_model.score(xtest,ytest)

#Regression Metrics/Cost Function
from sklearn.metrics import mean_absolute_error
mae_svc = mean_absolute_error(ytest, ans_svc) # (actual value - y) or (predicted - ans)

from sklearn.metrics import mean_squared_error
mse_svc = mean_squared_error(ytest,ans_svc)

rmse_svc = np.sqrt(mse_svc)

from sklearn.metrics import r2_score
r2_svc = r2_score(ytest,ans_svc)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines'],
    'Score_train': [acc_svc],
    'MAE': [mae_svc],
    'MSE': [mse_svc],
    'RMSE': [rmse_svc],
    'r2': [r2_svc]
                    })
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_model.fit(xtrain,ytrain)
ans_knn = knn_model.predict(xtest)
```

```python
acc_knn = knn_model.score(xtest,ytest)

#Regression Metrics/Cost Function
from sklearn.metrics import mean_absolute_error
mae_knn = mean_absolute_error(ytest, ans_knn) # (actual value - y) or (predicted - ans)

from sklearn.metrics import mean_squared_error
mse_knn = mean_squared_error(ytest,ans_knn)

rmse_knn = np.sqrt(mse_knn)

from sklearn.metrics import r2_score
r2_knn = r2_score(ytest,ans_knn)
```

```python
models.loc[len(models)] = {
    'Model': 'KNN',
    'Score_train': acc_knn,
    'MAE': mae_knn,
    'MSE': mse_knn,
    'RMSE': rmse_knn,
    'r2': r2_knn
}
```
